# Model Representation (Gradient Descent for Linear Regression)

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the cost
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = X * w + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

# Function to perform gradient descent
def gradient_descent(X, y, w, b, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for _ in range(iterations):
        predictions = X * w + b
        w_gradient = (1 / m) * np.sum((predictions - y) * X)
        b_gradient = (1 / m) * np.sum(predictions - y)

        w -= learning_rate * w_gradient
        b -= learning_rate * b_gradient

        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)

    return w, b, cost_history

# Generate some random data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Initialize parameters
w = np.random.randn()
b = np.random.randn()

# Set hyperparameters
learning_rate = 0.01
iterations = 1000

# Run gradient descent
w, b, cost_history = gradient_descent(X, y, w, b, learning_rate, iterations)

# Plot cost versus iterations
plt.plot(range(1, iterations + 1), cost_history, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost versus Iterations')
plt.show()

# Make predictions using the optimized parameters
predictions = X * w + b

# Plot the original data and the regression line
plt.scatter(X, y, label='Original Data')
plt.plot(X, predictions, color='red', label='Regression Line')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.show()
```
