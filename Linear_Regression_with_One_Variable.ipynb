# Model Representation

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the hypothesis function
def hypothesis(theta, X):
    """Hypothesis function for linear regression."""
    return np.dot(X, theta)

# Define the cost function
def cost_function(theta, X, y):
    """Compute the Mean Squared Error (MSE) cost for linear regression."""
    m = len(y)
    predictions = hypothesis(theta, X)
    squared_errors = (predictions - y) ** 2
    cost = (1 / (2 * m)) * np.sum(squared_errors)
    return cost

# Example usage
# Assuming you have a dataset X (features) and y (target variable)

# Add a column of ones to X for the bias term
X_with_bias = np.c_[np.ones((len(X), 1)), X]

# Initialize theta to zeros (or any other initial values)
theta_initial = np.zeros((X_with_bias.shape[1], 1))

# Calculate the initial cost
initial_cost = cost_function(theta_initial, X_with_bias, y)

# Print the initial cost
print(f'Initial Cost: {initial_cost}')

# Optionally, you can plot the data and regression line
plt.scatter(X, y, label='Training Data')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.title('Linear Regression with One Variable')
plt.legend()
plt.show()
```
